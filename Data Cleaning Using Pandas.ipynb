#  Data Cleaning In Pandas

```python
import pandas as pd
```

```python
df = pd.read_excel(r"C:\Users\Asus\OneDrive\Desktop\Python\pandas_dataset\DataCleaning_Using_Pandas\After Customer Call List.xlsx")
df
```

```python
# df.drop_duplicates()

df = df.drop_duplicates()
df
```

```python
# df.drop(columns = "Not_Useful_Column")

df = df.drop(columns = "Not_Useful_Column")
df
```

```python
# df["Last_Name"] = df["Last_Name"].str.lstrip("...")
# df["Last_Name"] = df["Last_Name"].str.lstrip("/")
# df["Last_Name"] = df["Last_Name"].str.rstrip("_")

# OR 

# df["Last_Name"] = df["Last_Name"].str.strip("123._/")
df["Last_Name"] = df["Last_Name"].str.strip("._/")
df
```

```python
#df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]','')
#                                OR
# df['Phone_Number'] = df['Phone_Number'].str.replace(r'[-/|]', '', regex=True)


# df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])

# df["Phone_Number"] = df["Phone_Number"].apply(lambda x: str(x))

# df["Phone_Number"] = df["Phone_Number"].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])

df["Phone_Number"] = df["Phone_Number"].str.replace('nan--','')

df["Phone_Number"] = df["Phone_Number"].str.replace('Na--','')

df
```

```python
# df["Address"].str.split(',',n=2, expand=True)
# df["Address"].str.split(',',n=1, expand=True)

df[["Street_Address", "State", "Zip_Code"]] = df["Address"].str.split(',',n=2, expand=True)
df
```

```python
df["Paying Customer"] = df["Paying Customer"].str.replace('Yes','Y')

df["Paying Customer"] = df["Paying Customer"].str.replace('No','N')

df
```

```python
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('Yes','Y')

df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('No','N')
df
```

```python
# df = df.replace('N/a','')
# df = df.replace('NaN','')


df=df.fillna('')
df
```

```python
for x in df.index:
    if df.loc[x, "Do_Not_Contact"] == 'Y':
        df.drop(x, inplace=True)

df
```

```python
for x in df.index:
#     if df.loc[x, "Phone_Number"] == '':
    if df.loc[x, "Phone_Number"] == '--':
        df.drop(x, inplace=True)
df

#          OR
#Another way to drop null values
#df = df.dropna(subset="Phone_Number"), inplace=True)
```

```python
# This will add index column at the end of the dataframe
# df['index'] = df.index

# This will help to drop the specified column
# df = df.drop('index', axis=1)

# This will help to add index column at the beginning of the dataframe
df.insert(0, 'index', df.index)
df
```

```python
df = df.set_index('index')
df
```

```python
df = df.reset_index(drop=True)
df
```
