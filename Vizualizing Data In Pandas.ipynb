# Pandas Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\Asus\OneDrive\Desktop\Python\pandas_dataset\Ice Cream Ratings.csv")
df = df.set_index('Date')
df
```

```python
# Here we are using matplotlib 

print(plt.style.available)

plt.style.use('classic')
```

```python
# df.plot()

df.plot(kind = 'line', title = 'Ice Cream Ratings', xlabel = 'Daily Ratings', ylabel = 'Scores')
```

```python
df.plot(kind = 'line', subplots = True)
```

```python
# df.plot(kind = 'bar')

df.plot(kind = 'bar', stacked = True)
```

```python
df['Flavor Rating'].plot(kind = 'bar', stacked = True)
```

```python
# df.plot.barh()

df.plot.barh(stacked = True)
```

```python
df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating', s = 300, c = 'Orange')
```

```python
# df.plot.hist()

df.plot.hist(bins = 20)
```

```python
df.boxplot()
```

```python
# df.plot.area()

df.plot.area(figsize = (10, 5))
```

```python
df.plot.pie(y = 'Flavor Rating', figsize = (10, 6))
```
